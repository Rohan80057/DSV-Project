```python
# ==============================================================================
# BLOCK 1 (Revised): Simple Data Loading & Verification
# ==============================================================================

# 1. Import the pandas library for data manipulation
import pandas as pd

# --- Action: Define the file path ---
# This is the full path to your dataset.
file_path = r"C:\Users\rohan das\OneDrive\Desktop\DSV_project\data\owid-covid-data.csv"

# --- Action: Load the Dataset ---
# We will load the CSV and use the 'parse_dates' argument to automatically
# convert the 'date' column to a proper datetime format.
try:
    df = pd.read_csv(file_path, parse_dates=['date'])
    print("✅ Dataset loaded successfully!")
    print(f"Shape of the dataset (rows, columns): {df.shape}")
except FileNotFoundError:
    print(f"❌ Error: The file was not found at the path: {file_path}")
    df = None  # Ensure df is None if loading fails

# --- Action: Verify the result ---
# If the dataframe was loaded, we check its structure and data types.
if df is not None:
    print("\n--- Verifying Data ---")
    
    # Display the first 5 rows to get a feel for the data
    print("First 5 rows:")
    print(df.head())
    
    # Display the DataFrame info, paying close attention to the 'date' column's Dtype.
    # We expect to see 'datetime64[ns]' for the 'date' column.
    print("\nDataFrame Info (checking data types):")
    df.info()
```

```python
# ==============================================================================
# BLOCK 2: DATA FILTERING, COLUMN SELECTION & HANDLING MISSING VALUES
# ==============================================================================

# --- Action: 1. Select Relevant Columns ---
# The dataset has many columns. Let's keep only the ones most relevant for
# a general analysis of trends, vaccinations, and demographics.
columns_to_keep = [
    'iso_code', 'continent', 'location', 'date', 'total_cases', 'new_cases',
    'total_deaths', 'new_deaths', 'total_cases_per_million',
    'total_deaths_per_million', 'total_vaccinations',
    'people_vaccinated', 'people_fully_vaccinated', 'population'
]

# Create a new DataFrame with only these columns
df_clean = df[columns_to_keep].copy()
print(f"Original shape: {df.shape}")
print(f"Shape after column selection: {df_clean.shape}")

# --- Action: 2. Remove Aggregate Locations ---
# The dataset includes rows for 'World', continents (e.g., 'Asia'), and income groups.
# We want to analyze only individual countries, so we must remove these aggregates
# to prevent double-counting and skewed results.
# We filter out rows where 'continent' is missing, as these are usually the aggregate rows.
df_countries = df_clean[df_clean['continent'].notna()].copy()

print(f"Shape after removing aggregate locations: {df_countries.shape}")
print(f"Number of unique countries remaining: {df_countries['location'].nunique()}")

# --- Action: 3. Handle Missing Values (Imputation) ---
# For daily 'new' data (cases, deaths), a missing value often means 0 reported that day.
# We will fill these specific NaNs with 0 to allow for accurate calculations.
cols_to_fill_zero = ['new_cases', 'new_deaths']
df_countries[cols_to_fill_zero] = df_countries[cols_to_fill_zero].fillna(0)

print("\n✅ Missing values for 'new_cases' and 'new_deaths' filled with 0.")

# --- Action: Final Check of Cleaned Data ---
print("\n--- cleaned data info ---")
df_countries.info()

print("\n--- First 5 rows of cleaned country data ---")
print(df_countries.head())
```

```python
# ==============================================================================
# BLOCK 3 (Corrected): DATA TRANSFORMATION & FEATURE ENGINEERING
# ==============================================================================

# Import numpy to handle infinite values
import numpy as np

# --- Action: 1. Extract Time-Based Features ---
df_countries['year'] = df_countries['date'].dt.year
df_countries['month'] = df_countries['date'].dt.month
print("✅ Added 'year' and 'month' columns.")

# --- Action: 2. Calculate Mortality Rate (Robustly) ---
# This is the ratio of total deaths to total cases.
# We perform a three-step process to prevent any warnings:
#   a. Calculate the rate. This may produce NaN (0/0) or inf (x/0).
#   b. Replace all infinite values (inf) with NaN.
#   c. Fill all remaining NaN values with 0.
df_countries['mortality_rate'] = (df_countries['total_deaths'] / df_countries['total_cases']) * 100
df_countries['mortality_rate'] = df_countries['mortality_rate'].replace([np.inf, -np.inf], np.nan).fillna(0)
print("✅ Calculated 'mortality_rate' and handled all NaN and infinite values.")

# --- Action: 3. Calculate Vaccination Rate ---
# This is the percentage of the population that has received at least one vaccine dose.
df_countries['vaccination_rate'] = (df_countries['people_vaccinated'] / df_countries['population']) * 100
print("✅ Calculated 'vaccination_rate' (People Vaccinated / Population).")


# --- Action: Final Check of Transformed Data ---
print("\n--- Transformed Data Info ---")
df_countries.info()

print("\n--- First 5 rows with new features ---")
print(df_countries.tail())

# --- Final Check: This should now run without any warnings ---
print("\n--- Descriptive statistics for new 'mortality_rate' column ---")
print(df_countries[['mortality_rate']].describe())
```

```python
# ==============================================================================
# BLOCK 4: EXPLORATORY DATA ANALYSIS (EDA) - PART 1: GLOBAL TRENDS
# ==============================================================================

# --- Action: 1. Import Visualization Libraries ---
import matplotlib.pyplot as plt
import seaborn as sns

# Set a visually appealing style for our plots
sns.set_style("darkgrid")
plt.rcParams['figure.figsize'] = (15, 8) # Set default figure size

print("✅ Visualization libraries imported and style set.")

# --- Action: 2. Analyze Global Trends Over Time ---
# To get the global daily numbers, we group our country data by 'date'
# and sum the 'new_cases' and 'new_deaths'.
global_trends = df_countries.groupby('date')[['new_cases', 'new_deaths']].sum().reset_index()

# --- Action: 3. Calculate 7-Day Rolling Average ---
# Daily data can be very noisy. A rolling average helps to see the underlying trend.
global_trends['new_cases_7day_avg'] = global_trends['new_cases'].rolling(window=7).mean()
global_trends['new_deaths_7day_avg'] = global_trends['new_deaths'].rolling(window=7).mean()
print("✅ Calculated global daily trends and 7-day rolling averages.")


# --- Action: 4. Create the Plot ---
# We will create a plot with two subplots (one for cases, one for deaths)
# to visualize the trends clearly.

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True) # 2 rows, 1 column

# Subplot 1: Global New Cases
ax1.plot(global_trends['date'], global_trends['new_cases_7day_avg'], label='7-Day Rolling Average', color='blue')
# We can also plot the raw daily data as a light background
ax1.bar(global_trends['date'], global_trends['new_cases'], label='Daily Cases', color='lightblue', alpha=0.6)
ax1.set_title('Global Daily COVID-19 Cases (7-Day Rolling Average)')
ax1.set_ylabel('Number of Cases')
ax1.legend()
# Format y-axis to be more readable (e.g., 5,000,000 instead of 5e6)
ax1.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))


# Subplot 2: Global New Deaths
ax2.plot(global_trends['date'], global_trends['new_deaths_7day_avg'], label='7-Day Rolling Average', color='red')
ax2.bar(global_trends['date'], global_trends['new_deaths'], label='Daily Deaths', color='lightcoral', alpha=0.6)
ax2.set_title('Global Daily COVID-19 Deaths (7-Day Rolling Average)')
ax2.set_xlabel('Date')
ax2.set_ylabel('Number of Deaths')
ax2.legend()
ax2.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))

# Display the plot
plt.tight_layout() # Adjusts plot parameters for a tight layout
plt.show()
```

```python
# ==============================================================================
# BLOCK 5 (Corrected): EDA - PART 2: COUNTRY-LEVEL COMPARISON
# ==============================================================================

# --- Action: 1. Get the Latest Data for Each Country ---
# To compare total figures, we need the most recent entry for each location.
# We sort the data by date and then take the last entry for each 'location'.
latest_data_per_country = df_countries.sort_values('date').drop_duplicates('location', keep='last')

print("✅ Extracted the most recent data for each country.")

# --- Action: 2. Identify Top 10 Countries by Cases and Deaths ---
# We'll sort the latest data to find the countries with the highest totals.
# We also filter out any countries with a very small population (e.g., < 1 million) to focus on major nations.
top_10_cases = latest_data_per_country[latest_data_per_country['population'] > 1000000].nlargest(10, 'total_cases')
top_10_deaths = latest_data_per_country[latest_data_per_country['population'] > 1000000].nlargest(10, 'total_deaths')

print("✅ Identified top 10 countries by total cases and deaths.")

# --- Action: 3. Create Bar Plots for Comparison ---
# We'll use subplots to display the two charts side-by-side.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8)) # 1 row, 2 columns

# Subplot 1: Top 10 Countries by Total Cases
# FIX: Added 'hue' and 'legend' parameters to align with modern Seaborn syntax
sns.barplot(
    x='total_cases',
    y='location',
    data=top_10_cases,
    ax=ax1,
    palette='viridis',
    hue='location',
    legend=False
)
ax1.set_title('Top 10 Countries by Total Confirmed Cases', fontsize=16)
ax1.set_xlabel('Total Cases')
ax1.set_ylabel('Country')
# Format x-axis labels to be readable (e.g., 100M instead of 100,000,000)
ax1.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1_000_000:.0f}M'))


# Subplot 2: Top 10 Countries by Total Deaths
# FIX: Added 'hue' and 'legend' parameters here as well
sns.barplot(
    x='total_deaths',
    y='location',
    data=top_10_deaths,
    ax=ax2,
    palette='plasma',
    hue='location',
    legend=False
)
ax2.set_title('Top 10 Countries by Total Deaths', fontsize=16)
ax2.set_xlabel('Total Deaths')
ax2.set_ylabel('') # No label needed as it's clear from the context
ax2.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1_000_000:.1f}M'))

# Display the plot
plt.suptitle('COVID-19 Impact: Absolute Counts by Country', fontsize=20)
plt.tight_layout(rect=[0, 0, 1, 0.96]) # Adjust layout to make room for suptitle
plt.show()
```

```python
# ==============================================================================
# BLOCK 8: EDA - PART 5: TIME-SERIES HEATMAP
# ==============================================================================

# --- Action: 1. Aggregate Data by Year and Month ---
# We group the data by year and month, summing the new cases to get a monthly total.
monthly_cases = df_countries.groupby(['year', 'month'])['new_cases'].sum().reset_index()

# --- Action: 2. Pivot the Data for the Heatmap ---
# We pivot the table so that years become the rows, months become the columns,
# and the cell values are the total new cases.
heatmap_data = monthly_cases.pivot(index='year', columns='month', values='new_cases')

# --- Action: 3. Create the Heatmap ---
plt.figure(figsize=(14, 8))
sns.heatmap(
    heatmap_data,
    cmap='Reds',           # Use a color scheme where darker means more cases
    annot=True,            # Annotate each cell with the number of cases
    fmt=".0f",             # Format annotations as integers
    linewidths=.5,         # Add lines between cells
    cbar_kws={'label': 'Total New Cases'} # Add a label to the color bar
)

plt.title('Global Monthly COVID-19 Cases Heatmap', fontsize=20)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Year', fontsize=14)
plt.show()
```

```python
# ==============================================================================
# BLOCK 10 (Corrected): SOCIO-ECONOMIC CORRELATION HEATMAP
# ==============================================================================

# --- Action: 1. Select Relevant Features for Correlation ---
# We define the list of socio-economic and health system factors we want to analyze,
# along with our key COVID-19 outcome metrics.
correlation_features = [
    'total_deaths_per_million', 'total_cases_per_million',
    'population_density', 'gdp_per_capita', 'extreme_poverty',
    'hospital_beds_per_thousand', 'life_expectancy'
]

# --- FIX: We now use the original, complete DataFrame 'df' ---
# 'df' contains all the columns from the original CSV. We get the latest
# data for each country from this complete source.
latest_full_data = df.sort_values('date').drop_duplicates('location', keep='last')

# Now, we select our desired columns from this complete, latest dataset.
# We also drop any rows that have missing values in any of these columns to ensure
# the correlation calculation is valid.
correlation_df = latest_full_data[correlation_features].dropna()
print(f"✅ Prepared data for correlation analysis. Analyzing {len(correlation_df)} countries with complete data.")

# --- Action: 2. Calculate the Correlation Matrix ---
correlation_matrix = correlation_df.corr()

# --- Action: 3. Plot the Correlation Heatmap ---
plt.figure(figsize=(12, 10))
sns.heatmap(
    correlation_matrix,
    annot=True,          # Show the correlation coefficients on the map
    cmap='coolwarm',     # Use a diverging colormap (blue=negative, red=positive)
    fmt=".2f",           # Format numbers to two decimal places
    linewidths=.5
)
plt.title('Correlation Between COVID-19 Outcomes and Socio-Economic Factors', fontsize=16)
plt.show()
```

```python
# ==============================================================================
# BLOCK 11: EDA - PART 8: CASE STUDY - COMPARING PANDEMIC WAVES
# ==============================================================================

# --- Action: 1. Select Countries and Prepare Data ---
case_study_countries = ['United States', 'India', 'Germany', 'Brazil', 'South Africa']
case_study_df = df_countries[df_countries['location'].isin(case_study_countries)].copy()

# We need to calculate the rolling average of new cases PER MILLION to compare fairly
case_study_df['new_cases_per_million_7day_avg'] = case_study_df.groupby('location')['new_cases'].rolling(window=7).mean().reset_index(0, drop=True) / (case_study_df['population'] / 1_000_000)
case_study_df = case_study_df.dropna(subset=['new_cases_per_million_7day_avg'])


# --- Action: 2. Plot the Time-Series Comparison ---
plt.figure(figsize=(18, 9))
sns.lineplot(
    data=case_study_df,
    x='date',
    y='new_cases_per_million_7day_avg',
    hue='location', # Create a different line for each country
    linewidth=2.5
)

plt.title('Pandemic Waves: Daily New Cases Per Million People (7-Day Avg)', fontsize=20)
plt.xlabel('Date', fontsize=14)
plt.ylabel('New Cases Per Million People', fontsize=14)
plt.legend(title='Country')
plt.grid(True, which="both", ls="--")
plt.show()
```

```python
# ==============================================================================
# BLOCK 12: EDA - PART 9: INTERACTIVE VISUALIZATION WITH PLOTLY
# ==============================================================================

# --- Action: 1. Import Plotly ---
import plotly.express as px

# --- Action: 2. Prepare Data (Using Case Study Countries) ---
# We can reuse the 'case_study_df' from the previous block.

# --- Action: 3. Create the Interactive Plot ---
# Plotly makes it incredibly simple to create a rich, interactive chart.
fig = px.line(
    case_study_df,
    x='date',
    y='new_cases_per_million_7day_avg',
    color='location',  # This is equivalent to 'hue' in seaborn
    title='Interactive View of Pandemic Waves (Daily New Cases Per Million)',
    labels={'new_cases_per_million_7day_avg': 'New Cases Per Million (7-Day Avg)', 'date': 'Date'}
)

# Update layout for a cleaner look
fig.update_layout(
    template='plotly_dark',
    legend_title_text='Country'
)

# Show the interactive plot
fig.show()
```

```python
# ==============================================================================
# BLOCK 13: INTRODUCTION TO PREDICTIVE MODELING
# ==============================================================================

# --- Action: 1. Import Scikit-learn ---
from sklearn.linear_model import LinearRegression
import numpy as np

# --- Action: 2. Prepare Data for a Single Country ---
country_to_model = 'United States'
model_df = df_countries[df_countries['location'] == country_to_model].copy()
model_df = model_df[['date', 'total_cases']].dropna()

# Create a numerical feature for time (days since the first case)
model_df['days_since_start'] = (model_df['date'] - model_df['date'].min()).dt.days

# --- Action: 3. Fit a Simple Linear Regression Model ---
# Prepare data for sklearn: X must be a 2D array, y a 1D array
X = model_df[['days_since_start']]
y = model_df['total_cases']

model = LinearRegression()
model.fit(X, y)

# --- Action: 4. Generate Predictions ---
model_df['predicted_cases'] = model.predict(X)

# --- Action: 5. Plot Actual vs. Predicted ---
plt.figure(figsize=(16, 8))
plt.scatter(model_df['date'], model_df['total_cases'], label='Actual Total Cases', alpha=0.5, s=10)
plt.plot(model_df['date'], model_df['predicted_cases'], color='red', linewidth=3, label='Simple Linear Trend Line')

plt.title(f'Simple Trend Model for Total Cases in {country_to_model}', fontsize=20)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Total Cases', fontsize=14)
plt.legend()
plt.grid(True)
print("NOTE: This plot shows a simple linear trend and does not capture the complex, non-linear nature of a pandemic.")
plt.show()
```

```python
# ==============================================================================
# BLOCK 15: DYNAMIC ANALYSIS - ROLLING CORRELATION
# ==============================================================================

# --- Action: 1. Prepare Global Trend Data ---
# We'll use the 'global_trends' DataFrame from Block 4.
# We need to ensure we're correlating the 7-day averages to smooth out noise.
rolling_corr_data = global_trends[['date', 'new_cases_7day_avg', 'new_deaths_7day_avg']].dropna()

# --- Action: 2. Calculate the Rolling Correlation ---
# We use a 90-day window to see how the correlation changes over quarterly periods.
rolling_correlation = rolling_corr_data['new_cases_7day_avg'].rolling(window=90).corr(rolling_corr_data['new_deaths_7day_avg'])

# --- Action: 3. Plot the Rolling Correlation ---
plt.figure(figsize=(18, 8))
rolling_correlation.plot(
    label='90-Day Rolling Correlation',
    color='purple',
    linewidth=3
)

# Add a horizontal line at 0 for reference
plt.axhline(0, color='grey', linestyle='--', linewidth=1)

plt.title('Dynamic Relationship: 90-Day Rolling Correlation Between Global Cases and Deaths', fontsize=20)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Correlation Coefficient', fontsize=14)
plt.ylim(-0.5, 1) # Set y-axis limits for better interpretation
plt.legend()
plt.grid(True, which='both', linestyle=':')
plt.show()
```

```python
# ==============================================================================
# BLOCK 16 (Revised): GEOSPATIAL ANALYSIS - ANIMATED CHOROPLETH MAP
# ==============================================================================

# --- Action: 1. Prepare Monthly Data for Animation (Robust Method) ---
# We need to aggregate our data by month and country.
# We will SUM the new cases and take the MEAN of the population for the month
# (population is mostly constant, so mean is a safe way to carry it over).
monthly_geo_data = df_countries.groupby(['iso_code', 'location', 'year', 'month']).agg({
    'new_cases': 'sum',
    'population': 'mean'
}).reset_index()

# Now, we calculate a true "Monthly New Cases Per Million"
monthly_geo_data['monthly_cases_per_million'] = (monthly_geo_data['new_cases'] * 1_000_000) / monthly_geo_data['population']

# Create a 'year_month' string for the animation frame
monthly_geo_data['year_month'] = monthly_geo_data.apply(lambda row: f"{int(row['year'])}-{int(row['month']):02d}", axis=1)
monthly_geo_data = monthly_geo_data.sort_values('year_month')

# We'll cap the color scale at a high percentile to avoid extreme outliers washing out the map colors.
color_max = monthly_geo_data['monthly_cases_per_million'].quantile(0.98)
print("✅ Monthly aggregated data for animation is ready using a robust calculation.")

# --- Action: 2. Create the Animated Choropleth with Plotly ---
fig = px.choropleth(
    monthly_geo_data,
    locations="iso_code",
    color="monthly_cases_per_million", # Use our new, correctly calculated column
    hover_name="location",
    animation_frame="year_month",      # This creates the animation slider
    color_continuous_scale=px.colors.sequential.Plasma,
    range_color=[0, color_max],        # Set a fixed color scale for consistency
    projection="natural earth",        # A nice map projection
    title="Evolution of Monthly Total New Cases Per Million by Country"
)

fig.update_layout(
    template='plotly_dark',
    coloraxis_colorbar_title_text='Cases Per Million' # Add a clear label to the color bar
)

fig.show()
```
